```python
import numpy as np
import pandas as pd

# --- Heuristic Function (Takes a single row, same as before) ---
def classify_player_role_heuristic(row):
    """
    Assigns a heuristic player role based on cumulative historical stats in the row.
    (Exact same implementation as before - designed to work on one row's cumulative stats)
    """
    # --- Thresholds (tuneable) ---
    WK_STUMPING_THRESHOLD = 0
    BOWLER_PERC_MATCHES_THRESH = 0.65
    BOWLER_AVG_BALLS_THRESH = 12
    BATTER_PERC_MATCHES_THRESH = 0.75
    BATTER_AVG_BALLS_THRESH = 10
    MODERATE_BOWLING_PERC = 0.25
    MODERATE_BOWLING_BALLS = 6
    MODERATE_BATTING_PERC = 0.30
    MODERATE_BATTING_BALLS = 5
    ALLROUNDER_CONTRIBUTION_RATIO = 1.5

    # --- Access Data Safely ---
    total_matches = row.get('Total_matches_played_sum', 0)

    # --- Handle Debutants (Based on cumulative matches before last match) ---
    if total_matches == 0: return 'Batter' # Default for low history

    # --- Get Cumulative Stats from the row ---
    stumpings = row.get('cumulative_Stumpings_sum', 0)
    innings_bowled = row.get('cumulative_Innings Bowled_sum', 0)
    balls_bowled = row.get('cumulative_Balls Bowled_sum', 0)
    innings_batted = row.get('cumulative_Innings Batted_sum', 0)
    balls_faced = row.get('cumulative_Balls Faced_sum', 0)

    # --- Identify Wicket Keeper ---
    if stumpings > WK_STUMPING_THRESHOLD: return "Wicket Keeper"

    # --- Calculate Significance Metrics (Based on cumulative before last match) ---
    perc_matches_bowled = innings_bowled / total_matches if total_matches > 0 else 0
    avg_balls_bowled = balls_bowled / total_matches if total_matches > 0 else 0
    perc_matches_batted = innings_batted / total_matches if total_matches > 0 else 0
    avg_balls_faced = balls_faced / total_matches if total_matches > 0 else 0

    # --- Classify based on Career Tendency ---
    is_significant_bowler = (perc_matches_bowled >= BOWLER_PERC_MATCHES_THRESH) or \
                            (avg_balls_bowled >= BOWLER_AVG_BALLS_THRESH and perc_matches_bowled > 0.5)
    is_significant_batter = (perc_matches_batted >= BATTER_PERC_MATCHES_THRESH) or \
                            (avg_balls_faced >= BATTER_AVG_BALLS_THRESH and perc_matches_batted > 0.5)

    is_moderate_bowler = (perc_matches_bowled >= MODERATE_BOWLING_PERC) or (avg_balls_bowled >= MODERATE_BOWLING_BALLS)
    is_moderate_batter = (perc_matches_batted >= MODERATE_BATTING_PERC) or (avg_balls_faced >= MODERATE_BATTING_BALLS)

    if is_significant_bowler and not is_moderate_batter: return 'Bowler'
    if is_significant_batter and not is_moderate_bowler: return 'Batter'

    if is_moderate_bowler and is_moderate_batter:
        bat_contribution = perc_matches_batted
        bowl_contribution = perc_matches_bowled
        if bowl_contribution > bat_contribution * ALLROUNDER_CONTRIBUTION_RATIO: return 'Bowling Allrounder'
        elif bat_contribution > bowl_contribution * ALLROUNDER_CONTRIBUTION_RATIO: return 'Batting Allrounder'
        else: return 'Allrounder'

    # --- Final Default ---
    if innings_bowled > 0 and innings_batted == 0 and perc_matches_bowled > 0.1: return 'Bowler'
    return 'Batter'

# --- Function to Assign HEURISTIC Roles (Overwrites existing column) ---
def assign_heuristic_player_roles(feature_df):
    """
    Assigns a player role to ALL players based purely on the heuristic
    applied to the cumulative stats from their LAST appearance in the feature_df.
    It creates or OVERWRITES the 'player_role' column.

    Args:
        feature_df (pd.DataFrame): The DataFrame with row-wise features, including
                                  'player_id', 'date', and cumulative stats.

    Returns:
        pd.DataFrame: DataFrame with a consistent 'player_role' column based on heuristic.
    """
    if 'player_id' not in feature_df.columns or 'date' not in feature_df.columns:
         print("Error: 'player_id' or 'date' column missing. Cannot find last occurrence.")
         return feature_df # Or raise error

    # --- 1. Find the last occurrence row for each player ---
    print("Finding last match occurrence for each player...")
    last_occurrence_df = feature_df.sort_values(['player_id', 'date'], ascending=[True, False]) \
                                   .drop_duplicates(subset='player_id', keep='first')

    if last_occurrence_df.empty:
        print("Warning: Could not determine last occurrences. Cannot assign roles.")
        return feature_df

    # --- 2. Apply Heuristic ONCE per player using stats from their last row ---
    print("Applying role heuristic based on last occurrence stats...")
    last_occurrence_df['heuristic_career_role'] = last_occurrence_df.apply(classify_player_role_heuristic, axis=1)

    # --- 3. Create Player-to-Heuristic-Role Mapping ---
    role_map = last_occurrence_df.set_index('player_id')['heuristic_career_role'].to_dict()

    # --- 4. Assign Heuristic Role to ALL Rows ---
    print(f"Assigning heuristic roles to all rows based on {len(role_map)} players...")
    # Use map to assign the determined role based on player_id to ALL rows.
    # This will create or overwrite the 'player_role' column.
    feature_df['player_role'] = feature_df['player_id'].map(role_map)

    # Check for players in feature_df not found in the map (shouldn't happen if map is based on feature_df)
    missing_map_count = feature_df['player_role'].isnull().sum()
    if missing_map_count > 0:
        print(f"Warning: {missing_map_count} rows could not be mapped to a role (player_id possibly missing from last occurrence map). Filling with 'Batter'.")
        feature_df['player_role'].fillna('Batter', inplace=True) # Fallback fill

    print("Finished assigning heuristic player roles.")
    print("Final Role Distribution:\n", feature_df['player_role'].value_counts())


    return feature_df

```

```python
# --- How to Apply ---
# Assuming 'feature_df' is your dataframe AFTER calculating all cumulative/rolling features
# Make sure it contains 'player_id', 'date', and all cumulative columns used by the heuristic.

main_df = assign_heuristic_player_roles(main_df)

# Now proceed with one-hot encoding
# role_dummies = pd.get_dummies(feature_df['player_role'], prefix='role', dummy_na=False)
# # Add player_role to the list of columns to drop from final X features if needed
# # ... prepare X for model ...
```
